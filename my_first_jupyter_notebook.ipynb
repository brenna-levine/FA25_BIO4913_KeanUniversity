Lists are mutable. We can change things in our list. I want to add elements to my list. I will use a method called append to do this.

```python
print(STUDENTS)
```

```python
#append method
STUDENTS.append('Brenna')
print(STUDENTS)
```

We can also remove elements from list. We remove them based on their index using a method called pop. It takes the format listname.pop(#)

```python
POPPED=STUDENTS.pop(9)
print(POPPED)
print(STUDENTS)
```

Sometimes you want to reverse the order of your list. This means put what was in the last index in the first index. To do this, we use the reverse() method.

```python
STUDENTS.reverse()
```

```python
BOLOGNA='oscar' #creates a new string variable
BOLOGNA_LIST=[] #creates an empty list
```

```python
print(BOLOGNA)
print(BOLOGNA_LIST)
```

```python
#loop through string to create a list
for CHAR in BOLOGNA: #for each character in variable bologna
        BOLOGNA_LIST.append(CHAR)
#lose your indent to get out of your loop
print(f"My bologna has a first name. It's {BOLOGNA_LIST}.")

```

Slicing is splitting a string into characters. You can index into a string just like you can index into a list (or nested list). We can slice one character, or many characters.

print(FIRST)
print('The LAST letter of my name is:', FIRST[5])

```python
#slicing many characters from a string at once
FULLNAME='Brenna Levine'
print(FULLNAME)
NEW_FIRST=FULLNAME[0:6] #characters 0 through 5, not inluding 6
NEW_LAST=FULLNAME[7:13] #characters 7 through 12, not including 13
print('First name:', NEW_FIRST)
print('Last name:', NEW_LAST)
```

We can slice lists just like we can slice strings. Slicing means pulling things out of strings or lists.

```python
#create a nested list
ELEMENTS=[['fluorine', 'F'],
          ['chlorine', 'Cl'],
          ['bromine', 'Br'],
          ['hydrogen', 'H'],
          ['helium', 'He']]
```

```python
print(ELEMENTS)
```

```python
#slice the last 3 lists in the nested list (in other words, see the last 3 lists)
ELEMENTS[-3:]
```
